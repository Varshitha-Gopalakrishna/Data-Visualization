```python
"""
Q1. Create a scatter plot to visualize the relationship between two variables, by generating a synthetic dataset
Answer-
"""
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#generating a synthetic dataset
np.random.seed(27)
x = np.random.rand(100)
y = 2 * x + np.random.normal(0, 0.1, 100)

#generating scatter plot
plt.figure(figsize = (8, 6))
sns.scatterplot(x = x, y = y, markers= 'x')

plt.title('Scatter Plot of Synthetic Data')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.show()
```

```python
"""
Q2) Generate a dataset of random numbers. Visualize the distribution of a numerical variable.
Answer-
"""
np.random.seed(42)
data = np.random.normal(loc = 0, scale = 1, size = 1000)

plt.figure(figsize = (8,6))
sns.histplot(data, kde= True)
plt.title('Distribution of Numerical Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
"""
Q3) Create a dataset representing categories and their corresponding values. Compare different categories based on numerical values.
Answer-
"""
import pandas as pd

df = {'Categories' : ['A', 'B', 'C', 'D'],
      'Values' : [27, 17, 14, 20]}

x = pd.DataFrame(df)

plt.figure(figsize = (8, 6))
sns.barplot(x = 'Categories', y = 'Values', data= x, hue = 'Categories')
plt.title('Bar Plot Comparison of Categories based on Values')
plt.show()
```

```python
"""
Q4) Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.
Answer-
"""
df = {'Category' : ['A', 'A', 'A', 'B', 'B', 'B', 'C', 'C', 'C', 'D', 'D', 'D'],
      'Value' : [27, 25, 24, 17, 12, 15, 20, 21, 22, 33, 35, 37]}

x = pd.DataFrame(df)

plt.figure(figsize = (8, 6))
sns.boxplot(x = 'Category', y = 'Value', data = x, hue = 'Category')
plt.title('Box Plot Distribution of Values Across Categories')
plt.show()
```

```python
"""
Q5) Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
Answer-
"""
np.random.seed(42)
size = 100

feature1 = np.random.rand(size)
feature2 = 2 * feature1 + np.random.normal(0, 0.1, size)
feature3 = 3 * feature1 - feature2 + np.random.normal(0,0.1, size)

#creating dataset
df = pd.DataFrame({'Feature1' : feature1, 'Feature2' : feature2, 'Feature3' : feature3})

#calculating correlation matrix
corr_matrix = df.corr()

#visualizing heatmap
plt.figure(figsize = (8, 6))
sns.heatmap(data = corr_matrix, annot= True)
plt.title('Correlation Matrix Heatmap')
plt.show()

```
